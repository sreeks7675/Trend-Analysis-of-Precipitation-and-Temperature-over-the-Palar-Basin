```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
file=input("Enter file name: ")
df=pd.read_csv(file)
max_annual_rain=df.groupby("Year")["RF in mm"].max()
```

```python
plt.xticks(range(min(max_annual_rain.index), max(max_annual_rain.index) + 1))
plt.plot(max_annual_rain)
plt.xlabel('Year')
plt.ylabel('Maximum Rainfall (mm)')
plt.title('Maximum Annual Rainfall Plot')

if len(max_annual_rain.index) > 10:
    plt.xticks(rotation=45)

plt.show()
```

```python
length=max_annual_rain.size
max_val=max_annual_rain.max()
first_half=max_annual_rain.iloc[0:length//2]
second_half=max_annual_rain.iloc[length//2:]
yearF=max_annual_rain.index[0]
yearS=max_annual_rain.index[length//2]
x=[]
y=[]
#print(int(str(dict(yearF).keys())))
#['Year']
for i in range(0,length//2):
   #plt.plot(first_half[yearF+i],second_half[yearS+i])
    x.append(first_half[yearF+i])
    y.append(second_half[yearS+i])
    print(first_half[yearF+i],second_half[yearS+i])
plt.xlim(0, max_val+50)
plt.ylim(0, max_val+50)
plt.plot()
x.sort()
y.sort()
plt.scatter(x, y, color= "blue", s=30)
xpoints = np.array([0, max_val+50])
ypoints = np.array([0, max_val+50])

plt.plot(xpoints, ypoints,color="red")
# x-axis label
plt.xlabel(str(yearF)+"-"+str(yearF+length//2))
plt.ylabel(str(yearS)+"-"+str(yearS+length//2))
# frequency label
#plt.ylabel('y - axis')
# plot title
plt.title('ITA for '+file)
#completeDf=first_half['RF in mm']+second_half['RF in mm']
#completeDf
#second_half.describe
```

```python
countP=0
countN=0
countE=0
for i in range(0,length//2):
    if(x[i]>y[i]):
        countN+=1
    elif(x[i]<y[i]):
        countP+=1
    else:
        countE+=1

if(countP>countN):
    print("Positive/Increasing trend is observed")
elif(countP<countN):
    print("Negative/decreasing trend is observed")
elif((countP==countN) or (countP==0 and countN==0)):
    print("No trend observed")
```

```python
import sklearn.linear_model as skl
avg_annual_rain=df.groupby("Year")["RF in mm"].mean()
print(avg_annual_rain)

plt.xticks(range(min(avg_annual_rain.index), max(avg_annual_rain.index) + 1))
plt.plot(avg_annual_rain)
plt.xlabel('Year')
plt.ylabel('Average Rainfall (mm)')
plt.title('Average Annual Rainfall Plot')

if len(max_annual_rain.index) > 10:
    plt.xticks(rotation=45)

plt.show()
```

```python
model = skl.LinearRegression()

# Fit the model to your data (x and y)
year=np.array([i for i in range(1993,2023)])
model.fit(year.reshape(-1,1), avg_annual_rain)  # Reshape x for 2D compatibility

# Get the slope and intercept coefficients
slope = model.coef_[0]
intercept = model.intercept_

# Predict new y values for a given x value
new_x = 2023
predicted_y = model.predict([[new_x]])

print("Slope:", slope)
print("Intercept:", intercept)
print("Predicted y for", new_x, ":", predicted_y)
```
